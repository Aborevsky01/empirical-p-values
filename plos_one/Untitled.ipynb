```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
train_negative = np.random.normal(loc=-8.5, scale=4.0, size=[5000])
train_positive = np.random.normal(loc=8.5, scale=4.0, size=[5000])


test_positive = np.random.normal(loc=8.5, scale=4.0, size=[1000])
test_negative = np.random.normal(loc=-8.5, scale=4.0, size=[5000])
train = np.concatenate((train_positive, train_negative))
test = np.concatenate((test_positive, test_negative))
```

```python
train.shape
```

```python
fig, ax = plt.subplots(figsize=(16,8))
a=sns.histplot(train, bins=np.arange(-30,30,0.3), color='purple', kde=True, fill=True)
a=sns.histplot(test, bins=np.arange(-30,30,0.3), color='yellow', kde=True, fill=True)
a=sns.histplot(train_positive, bins=np.arange(-30,30,0.3), color='red', kde=True, fill=True)
a=sns.histplot(train_negative, bins=np.arange(-30,30,0.3), color='blue', kde=True, fill=True)
a=sns.histplot(test_positive, bins=np.arange(-30,30,0.3), color='green', kde=True, fill=False)
a=sns.histplot(test_negative, bins=np.arange(-30,30,0.3), color='orange', kde=True, fill=False)
f = plt.figure()
plt.show()
f.savefig("d:\\distributinos.png", bbox_inches='tight')
```
